```python
import utils as u
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="darkgrid")
```

```python
data = u.prepare_data("SIM-01-26-2022-15:29.log")
```

```python
ud = u.update_delay(data,"1")
kwargs = {'cumulative': True}
f, (ax_normal, ax_log) = plt.subplots(ncols=2, nrows=1, sharey=True, gridspec_kw={'wspace':0.05})
ax_normal.axvline(100, ls='--', color='r')
ax_log.set(xscale="log")

b = sns.distplot(ud,ax=ax_normal , kde_kws=kwargs)
a = sns.distplot(ud,ax=ax_log , kde_kws=kwargs)

b.set(xlim=(0,150))
a.set(xlim=(150,None))

# sns.despine(ax=ax_normal)
# sns.despine(ax=ax_log)


ax = ax_log
d = .02  # how big to make the diagonal lines in axes coordinates
# arguments to pass to plot, just so we don't keep repeating them
kwargs = dict(transform=ax.transAxes, color='k', clip_on=False)
ax.plot((-d, +d), (-d, +d), **kwargs)        # top-left diagonal

ax2 = ax_normal
kwargs.update(transform=ax2.transAxes)  # switch to the bottom axes
ax2.plot((1-d, 1+d), (- d,  d), **kwargs)  # bottom-left diagonal



plt.show()
```
